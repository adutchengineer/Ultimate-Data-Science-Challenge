## Exploratory data analysis

```python
#First step. Let's read in the json file
import json

with open('logins.json') as json_data:
    d = json.load(json_data)
```

```python
#let's look at what the data is like
import pandas as pd
df = pd.DataFrame.from_dict(d)
df = df.dropna()
```

```python
import datetime as dt
df['login_time'] = df['login_time'].apply(lambda x: 
                                    dt.datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
df['min'] = df['login_time'].dt.minute
df['hour'] = df['login_time'].dt.hour
df['day'] = df['login_time'].dt.day
df['week'] = df['login_time'].dt.week
df['weekday'] = df['login_time'].dt.dayofweek
```

```python
df.head()
```

```python
min_interval = df.set_index('login_time').resample('15min').size()
print(min_interval.head())
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1,figsize=(7,5))
min_interval.plot()
plt.xlabel('Date')
plt.ylabel('Sum of Logins in Every 15 Minutes')
plt.title('Sum of Logins for Every 15 minutes from January to April, 1970')
plt.show()
```

A few things to note already from this plot. The overall date range is from January to April in 1970. There is definitely some periodic pattern going on.

```python
# Dropping date-time features for predictive modeling
df.drop('last_trip_date', axis=1, inplace=True,errors='ignore')
df.drop('signup_date', axis=1, inplace=True,errors='ignore')
# Saving 'retained' values for later
y = df['retained'].values
# Dropping 'retained' values
df = df.drop('retained', axis=1, errors='ignore')
df
```

```python
df_encoded = pd.get_dummies(df, columns = ['city', 'phone', 'ultimate_black_user'])
df_encoded.head()
```

```python
df.info()
```

```python
X = df_encoded.values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Importing TensorFlow and the Keras libraries with packages
import tensorflow
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Importing packages and building new classifer with cross validation
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 7, kernel_initializer = 'uniform', activation = 'relu', input_dim = 14))
    classifier.add(Dense(units = 7, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier

# Fitting Neural Network to Training set and getting accuracies
classifier = KerasClassifier(build_fn = build_classifier, batch_size = 16, epochs = 10)
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 5)
```

```python
mean = accuracies.mean()
variance = accuracies.std()
print('Average: ', ('%.4f' % mean))
print('Standard Deviation: ', ('%.4f' % variance))
```

```python
# Importing packages and building new classifer with grid search
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense

# Building our Neural Network with empty optimizer, batch size, and epochs
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units = 7, kernel_initializer = 'uniform', activation = 'relu', input_dim = 14))
    classifier.add(Dense(units = 7, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
classifier = KerasClassifier(build_fn = build_classifier)

# Parameters for batch size, epochs, and optimizer functions
parameters = {'batch_size': [10, 20],
              'epochs': [10,50],
              'optimizer': ['adam', 'rmsprop']}

# Setting up Grid Search
grid_search = GridSearchCV(estimator = classifier,
                          param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 5)
grid_search = grid_search.fit(X_train, y_train)
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
