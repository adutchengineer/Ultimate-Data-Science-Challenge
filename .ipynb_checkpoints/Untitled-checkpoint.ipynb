```python
def check_negatives(string):
    #Check for negative values so it does not interfere with the subtractioncharacter
    temp_list = list(string.replace(" ",""))
    for i in range(len(temp_list)):
        if (temp_list[i] in ops and temp_list[i+1] in ops) or \
        (temp_list[i] == '(' and temp_list[i+1] in ops):
            try: 
                value = ''.join(temp_list[i+1:i+3])
                if int(value) < 0:
                    temp_list[i+1] = temp_list[i+1].replace(temp_list[i+1],'!')
            except:
                if int(temp_list[i+1]+str(1)) < 0:
                    temp_list[i+1] = temp_list[i+1].replace(temp_list[i+1],'!')
    return ''.join(temp_list)

def bracket(string):
    #Add brackets
    return "(" + string + ")"
    
def isnum (string):
    return all([i not in string for i in ops])

def bracket_removal(infix):
    stack = []
    postfix = []
    numstack = []
    opstack = []
    negative_values = None
    global ops  #made it global so that other functions could use them as well
    ops = {"+": 1,"-": 1,"*": 2,"/": 2} #operators along with their values for precedence
    infix_unaltered = infix
    infix=check_negatives(infix) #change the negative sign to a "!" 
    
    if '!' in infix:
        indices = [i  for i in range(0,len(infix)) if infix[i] == '!']
        negative_values = [infix[index+1] for index in indices]
        infix = infix.replace('!','') #delete the negative value
        negative_signs = list(zip(indices,negative_values))
    
    for c in infix:
        if c in ops:
            if stack:
                top = stack[-1] #Grabs top of the stack
                while top in ops and ops[top] >= ops[c]:
                    postfix.append(stack.pop())
                    if stack:
                        top = stack[-1]
                    else:
                        break
            stack.append(c)
        elif c in ['(',')']:
            if c == ")":
                while stack[-1] != "(":
                    postfix.append(stack.pop())
                stack.pop()  # Removes ( from the top of the stack
            else:
                stack.append(c)  # c == '('
        else:
            postfix.append(c)
            
    postfix.extend(token for token in reversed(stack) if token in ops)
    inputpfs = ''.join(postfix) #make it a string again
    
    for i in inputpfs:
        if i not in ops:
            numstack.append(i)
        else:
            right = numstack.pop()
            left = numstack.pop()
            if not isnum(right):
                if i in ["/", "-"]:
                    right = bracket(right)
                elif i=="*" and ("-" in right or "+" in right):
                    right = bracket(right)
            newresult = "".join([left, i, right])
            numstack.append(newresult)
    
    if negative_values:
        numstack = [x for x in numstack[0]] #flatten out the string in list
        for value in negative_signs:
            for i in range(0,value[0]-1):
                if numstack[i] in ops and numstack[i+1] == value[1]:
                    numstack[i+1] = "-"+value[1]
                elif numstack[i] in ops and numstack[i+2] == value[1]:
                    numstack[i+2] = "-"+value[1]
                    
    if ' ' in infix_unaltered:
        return " ".join(numstack)
    else:
        return "".join(numstack)
```

```python
tests = [
   "1*(2+(3*(4+5)))",
   "2 + (3 / -5)",
  "x+(y+z)+(t+(v+w))",
  "((y+z))+a",
   "(a-b)-a",
   "a-(a-b)",
   "a+(a-b)",
    "((a+a+a))",
    "1*(-2+(-3*(-4+5)))",
    "a + (b / -a)"
]

for test in tests:
    print ("Testing '%s': %s" % (test, bracket_removal(test)))
```
